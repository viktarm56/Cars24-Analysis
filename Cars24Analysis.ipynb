### Избавимся от дубликатов в данных

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
```

Датасет не содержит явных дубликатов. Проверим также на неявные:

```python
# Импортируем библиотеку FuzzyWuzzy для поиска неявных дубликатов в названиях моделей автомобилей
from rapidfuzz import process, fuzz

df['model_name'] = df['model_name'].str.strip().str.lower()
df['price_rounded'] = df['price'].round(-3)
df['km_driven_rounded'] = df['km_driven'].round(-3)

# Функция для поиска неявных совпадений
""" 
Цикл, используя методы библиотеки rapidfuzz, создает список "канонических" названий - общегрупповых.
Он перебирает каждое уникальное название из датасета и создает список и словарь: спислок 
для канонических названий и итоговый словарь, в котором ключи - названия из датасета, а значения - 
канонические общегрупповые названия. Значение из датасета присоединяется к общегрупповому названию,
если метод fuzz.WRatio показывает схожесть свыше 90%
"""

canonical_names_list = []
name_matched = {}

names_list = df['model_name'].unique().tolist()
for name in names_list:
    if not canonical_names_list:
        canonical_names_list.append(name)
        name_matched[name] = name
    else:
        match = process.extractOne(name, canonical_names_list, scorer=fuzz.WRatio)
        if match and match[1] >= 80:
            name_matched[name] = match[0]
        else:
            canonical_names_list.append(name)
            name_matched[name] = name

# Создаем новый столбец для схожих названий
df['matched_model_name'] = df['model_name'].map(name_matched)

# Выведем строки с одинаковым общим названием и другими параметрами, вероятнее всего это дубликаты
print(df.duplicated(subset=['matched_model_name', 'price_rounded', 'manufacturing_year', 'km_driven_rounded']).sum())
duplicates = df[df.duplicated(subset=['matched_model_name', 'price_rounded', 'manufacturing_year', 'km_driven_rounded'], keep = False)]
duplicates
```

Тяжело сделать однозначный вывод о том, что эти строки являются дубликатами: у них отличаются значения нюансов и перекрашенных элементов. Это может объясняться как перевыставлением одного автомобиля, так и выставленными двумя разными. 

В целом, учитывая небольшой размер исследуемого датасета, первый вариант выглядит вероятнее. В любом случае доля удаленных значений будет незначительной (0,015%) и не должна повлиять на дальнейшее исследование.

```python
df = df.drop_duplicates(subset=['matched_model_name', 'price_rounded', 'manufacturing_year', 'km_driven_rounded'], keep = 'last')
```

### 1. Гипотеза H₀: Между автомобилями с дизельным и бензиновым типами топлива нет статистически значимой разницы в стоимости.
   
    Проведем U-тест Манна-Уитни для проверки гипотезы о том, что дизельные автомобили в среднем дороже бензиновых. Предварительно ограничим выборку одним ценовым диапазоном, чтобы исключить влияние иных факторов, таких как класс автомобиля.

```python
price_range = (df['price'].between(300000, 800000))

# Фильтруем по типу топлива и ценовому диапазону, затем берем случайную выборку
petrol_prices = df[(df['fuel_type'] == 'Petrol') & price_range]['price'].sample(n=50, random_state=42)
diesel_prices = df[(df['fuel_type'] == 'Diesel') & price_range]['price'].sample(n=50, random_state=42)

# Проверяем размеры групп
print("Размер выборки бензиновых автомобилей:", petrol_prices.count())
print("Размер выборки дизельных автомобилей:", diesel_prices.count())

# Выполняем U-тест Манна-Уитни
stat, p = stats.mannwhitneyu(petrol_prices, diesel_prices, alternative='less')
print("p-value:", p)

if p < 0.05:
    print("Отвергаем H₀: Дизельные автомобили в среднем дороже бензиновых (p < 0.05)")
else:
    print("Не отвергаем H₀: Нет значимой разницы в средней цене (p >= 0.05)")
```

<div style="background-color: #E6FFE6; color: black; padding: 15px; border-radius: 8px; border: 1px solid #C8E6C9; margin-bottom: 15px;">
    <h4>Нулевая гипотеза отвергнута!</h4>
    <p>Проведенный статистический тест однозначно подтвердил наличие статистически значимых различий в стоимости автомобилей. Таким образом, альтернативная гипотеза о том, что дизельные автомобили в среднем дороже бензиновых, подтверждена.</p>
</div>

### 2. Гипотеза H₀: Количество предыдущих владельцев не влияет на стоимость.
   
    Для нахождения статистически важной разницы между группами используем непараметрический тест Крускала-Уоллиса, подходящий для данных с ассиметричным распределением

```python
owners_1 = df[df['ownership'] == 1]['price']
owners_2 = df[df['ownership'] == 2]['price']
owners_3 = df[df['ownership'] == 3]['price']

stat, p = stats.kruskal(owners_1, owners_2, owners_3)
print("p-value:", p)

if p < 0.05:
    print("Отвергаем H₀: Число владельцев влияет на цену (p < 0.05)")
else:
    print("Не отвергаем H₀: Нет значимой разницы в средней цене (p >= 0.05)")
```

Теперь при помощи U-теста Манна-Уитни проверим существует ли разница между каждой парой групп:

```python
stat_12, p_12 = stats.mannwhitneyu(owners_1, owners_2)
print("1 vs 2, p-value:", p_12)
if p_12 < 0.05 / 3:
    print("Отвергаем H₀: Существуем значимая разница цены для автомобилей с одним и двумя владельцами")
else:
    print("Не отвергаем H₀: Не существует статистически значимой разницы цены для автомобилей с одним и двумя владельцами")
    
stat_13, p_13 = stats.mannwhitneyu(owners_1, owners_3)
print("1 vs 3, p-value:", p_13)
if p_13 < 0.05 / 3:
    print("Отвергаем H₀: Существуем значимая разница цены для автомобилей с одним и тремя владельцами")
else:
    print("Не отвергаем H₀: Не существует статистически значимой разницы цены для автомобилей с одним и тремя владельцами")

stat_23, p_23 = stats.mannwhitneyu(owners_2, owners_3)
print("2 vs 3, p-value:", p_23)
if p_23 < 0.05 / 3:
    print("Отвергаем H₀: Существуем значимая разница цены для автомобилей с двумя и тремя владельцами")
else:
    print("Не отвергаем H₀: Не существует статистически значимой разницы цены для автомобилей с двумя и тремя владельцами")
```

<div style="background-color: #E6FFE6; color: black; padding: 15px; border-radius: 8px; border: 1px solid #C8E6C9; margin-bottom: 15px;">
    <h4>Нулевая гипотеза отвергнута!</h4>
    <p>Проведенный статистический тест однозначно подтвердил наличие статистически значимых различий. Альтернативная гипотеза о том, что число владельцев влияет на цену, подтверждена.</p>
</div>

### 3. Гипотеза H₀: Наличие запасного ключа не оказывает влияние на стоимость
    Используем U-тест Манна-Уитни для проверки:

```python
with_key = df[df['spare_key'] == True]['price']
without_key = df[df['spare_key'] == False]['price']

print(f'Количество значений с запасным ключом: {with_key.count()}')
print(f'Количество значений без запасного ключа:{without_key.count()}')
print()
stat, p = stats.mannwhitneyu(with_key, without_key)
print("Гипотеза 1: Влияние запасного ключа")
print("p-value:", p)

print()

if p < 0.05:
    print("Отвергаем H₀: Наличие запасного ключа влияет на цену (p < 0.05)")
else:
    print("Не отвергаем H₀: Нет значимого влияния запасного ключа на цену (p >= 0.05)")

# Медианы для понимания направления
print("Медиана с ключом:", with_key.median())
print("Медиана без ключа:", without_key.median())
```

<div style="background-color: #E6FFE6; color: black; padding: 15px; border-radius: 8px; border: 1px solid #C8E6C9; margin-bottom: 15px;">
    <h4>Нулевая гипотеза отвергнута!</h4>
    <p>Результаты статистического теста указывают на то, что наличие запасного ключа действительно оказывает влияние на среднюю стоимость автомобиля. Более того, как показывают медианные значения, запасной ключ положительно влияет на стоимость, делая автомобили с ним в среднем дороже.</p>
</div>

### 4. Гипотеза H₀: Пробег автомобиля не оказывает статистически важное влияние на его стоимость.
   
    Ранее мы уже находили значение корреляции Спирмана для значения пробега и получили довольно низкое значение. С помощью статистической гипотезы мы можем дополнить наш предыдущий вывод.

```python
corr, p = stats.spearmanr(df['km_driven'], df['price'])
print("\nГипотеза 2: Влияние пробега")
print("Коэффициент Спирмена:", corr)
print("p-value:", p)

if p < 0.01:
    print("Отвергаем H₀: Пробег влияет на цену (p < 0.05)")
else:
    print("Не отвергаем H₀: Нет значимой связи между пробегом и ценой (p >= 0.05)")
```

Однако, рассчитанный коэффициент корреляции Спирмена (-0.29) указывает на то, что, хотя эта связь и существует, пробег не является определяющим фактором ценообразования. Это означает, что другие характеристики, такие как год выпуска автомобиля и объем двигателя, вероятно, оказывают более существенное влияние на конечную стоимость.</p>

    <h2 style="margin: 0; color: black;">5. Проведем кластеризацию данных и проанализируем полученные кластеры</h2>

Для получения дополнительных инсайтов о ценообразовании на рынке подержанных автомобилей проведем кластеризацию по следующим признакам: Год выпуска, пробег, объем двигателя, цена. 

Кластеризация поможет сегментировать автомобили на группы для анализа влияния признаков на цену.

```python
df.columns
```

```python
# Импортируем необходимый модуль из библиотеки sklearn
from sklearn.preprocessing import StandardScaler

X = df[['price', 'manufacturing_year', 'engine_capacity', 'km_driven']]

# Стандартизируем данные для кластеризации при помощи k-value
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Найдем оптимальное количество кластеров при помощи Метода локтя:

```python
# Импортируем необходимый модуль 
from sklearn.cluster import KMeans

# Найдем значения инерции для разного количества кластеров
inertia = []
for k in range(1, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Построим график для метода локтя
plt.figure(figsize= (4, 2))

plt.plot(range(1, 7), inertia, marker='o')
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Значение инерции')

plt.show()
```

Оптимальным значением количества кластеров будет 4, дальнейшее добавление кластеров малоэффективно.

Добавим в исходный датасет столбец со значениями кластеров и выведем средние значения, чтобы понять их содержимое:

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

selected_cols = ['price', 'manufacturing_year', 'engine_capacity', 'km_driven']

df.groupby('cluster')[selected_cols].agg({
    'price': ['count', 'mean'],
    'manufacturing_year': 'mean',
    'engine_capacity': 'mean',
    'km_driven': 'mean'
                                         })
```

При помощи диаграммы параллельных координат наглядно изобразим различия в кластерах (Данные по оси Y нормализованы):

```python
# Добавим текстовые столбцы с названиями кластеров
df['cluster_name'] = df['cluster'].astype('str')
replace_dict = {
    0: 'Малобюджетные старые',
    1: 'Среднебюджетные новые',
    2: 'Высокобюджетные новые',
    3: 'Среднебюджетные старые'
}
df['cluster_name'] = df['cluster'].replace(replace_dict)
```

```python
from pandas.plotting import parallel_coordinates

parallel_data = df[['price', 'km_driven', 'manufacturing_year', 'engine_capacity', 'cluster_name']]

# Нормализуем данные
parallel_data_normalized = parallel_data.copy()
for col in ['price', 'km_driven', 'manufacturing_year', 'engine_capacity']:
    parallel_data_normalized[col] = (parallel_data[col] - parallel_data[col].min()) / (parallel_data[col].max() - parallel_data[col].min())

# Для большей наглядности рассчитаем средние значения после нормализации
parallel_data_normalized_grouped = parallel_data_normalized.groupby('cluster_name')\
[['price', 'km_driven', 'manufacturing_year', 'engine_capacity']].mean().reset_index()

plt.figure(figsize=(6, 4))
parallel_coordinates(parallel_data_normalized_grouped, 'cluster_name', colormap='viridis')
plt.title('Признаки по кластерам (Данные нормализованы и усреднены)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    borderaxespad=0
)

plt.savefig('kval_parallel_data.png')
plt.show()
```

Построим диаграмму Паретто, чтобы оценить доли кластеров в общей сумме выставленных на продажу автомобилей в датасете:

```python
cluster_price = df.groupby('cluster_name')['price'].sum().sort_values(ascending=False)

cumulative_percentage = cluster_price.cumsum() / cluster_price.sum() * 100

# Создаём диаграмму Парето
fig, ax1 = plt.subplots(figsize=(10, 6.8))

ax1.bar(cluster_price.index, cluster_price, color='yellow', alpha=0.7)
ax1.set_xlabel('Кластер')
ax1.set_ylabel('Суммарная стоимость (рупии)', color='C0')
ax1.tick_params(axis='y', labelcolor='C0')
plt.xticks(rotation=15)

for i, v in enumerate(cluster_price):
    ax1.text(i, v + 0.02 * v, f'{int(v)}', ha='center', va='bottom')

ax2 = ax1.twinx()
ax2.plot(cluster_price.index, cumulative_percentage, color='black', marker='o', linewidth=2)
ax2.set_ylabel('Кумулятивный процент (%)', color='C1')
ax2.tick_params(axis='y', labelcolor='C1')
ax2.set_ylim(0, 100)

for i, v in enumerate(cumulative_percentage):
    ax2.text(i, v + 2, f'{v:.1f}%', ha='center', va='bottom')

ax2.grid(True, linestyle='--', alpha=0.7)

plt.title('Диаграмма Парето: Суммарная стоимость по кластерам')
plt.tight_layout()
plt.savefig('paretto.png')
plt.show()
```

Среднебюджетные автомобили составляют 65.8% рынка автомобилей Индии (по совокупной стоимости)

Теперь сравним корреляционные матрицы для каждого из кластеров. Будем использовать незадействованные в кластеризации столбцы с нюансами и перекрашенными частями, а также наличием ключа:

### Расссмотрим распределние автомобилей по типу топлива и трансмиссии внутри кластеров

Чтобы лучше понять рынок подержанных автомобилей, рассматрим распределение автомобилей по типу топлива и трансмиссии внутри кластеров:

1. Тип топлива

```python
fuel_type_dist = df.groupby('cluster_name')['fuel_type'].value_counts(normalize=True).unstack().fillna(0)
print("Доли fuel_type по кластерам:")
print(fuel_type_dist)
```

Автомобили с бензиновым двигателем доминируют во всех категориях - это наиболее популярный тип топлива для автомобилей. Альтернативами выступают дизельное топливо и природный газ.

- Автомобили на газу составляют очень малую долю среди высокобюджетных и среднебюджетных старых, при этом более заметную долю среди малобюджетных старых и среднебюджетных новых.
- Автомобили с дизельными двигателями приходятся практически исключительно на категорию среднебюджетных старых автомобилей. Можно говорить о том, что производители отказываются от этого типа топлива, вероятно, из-за его неэкологичности.

**При помощи теста Манна-Уитни проверим положительное влияние дизельного двигателя на стоимость:**

```python
cluster_3_data = df[df['cluster_name'] == 'Среднебюджетные старые']

petrol_prices = cluster_3_data[cluster_3_data['fuel_type'] == 'Petrol']['price']
diesel_prices = cluster_3_data[cluster_3_data['fuel_type'] == 'Diesel']['price']

stat, p = stats.mannwhitneyu(petrol_prices, diesel_prices, alternative='less')
print(f"p-value для гипотезы 'Petrol < Diesel' в кластере 'Старые комфортные': {p:.4f}")
if p < 0.05:
    print("Гипотеза подтверждается: дизельные машины дороже бензиновых (p < 0.05).")
else:
    print("Гипотеза не подтверждается: разница в ценах не значима (p >= 0.05).")
```

**Наш более ранний вывод подтвержден.**
